인공지능 기술들에 대해서 분석해 보고, 나의 흥미를 끄는 기술들은 무엇이 있는지 찾아보기

## 1. 언어
- TensorFlow를 활용한 '수화를 텍스트로 바꿔주는 인공지능'
  - 청각장애인은 듣지 못하고, 많은 사람들은 수화를 알지 못한다. 이 간극을 해결하는 인공지능.
  - 실시간 동영상 캡션
  - 시퀀스 네트워크를 위한 뉴럴 네트워크


## 2. 음성
- 화자가 사용하는 단어와 목소리 톤 등을 분석해 감정까지 잡아내는 인공지능
  - 마케팅 및 서비스 제공, 범죄 등 다양한 범위에서 사용할 수 있다.
  - STT(Speech To Text) : 음성 입력 과정을 거쳐 사람의 음성을 컴퓨터가 이해할 수 있도록 텍스트화
  - NLP(Natural Language Processing) : 자연어 처리 과정을 통해 컴퓨터가 문장에 담긴 의도를 파악하게 한다.
- 사투리를 사용하는 대화형 인공지능
  - 같은 한국어라도 지역색이 짙은 사투리는 단번에 알아듣기 힘들다.
  - 사투리를 사용한 일상대화를 빅데이터로 만드는 사업
  - AI 콜 서비스 
  - LG전자 '씽큐': 지역 카테코리 별로 분류한 음성 데이터를 이용해 지역별 음성 특징을 학습한다.
      음성 데이터가 쌓일수록 정확도는 높아진다.
      액센트, 지역별 특징 단어와 상기 단어의 사용 횟수 등을 이용해 인공지능이 지역 카테고리를 예측한다. 예측에 따라 가장 높은 점수를 획득한 지역 카테고리로 음성일 인식하고 소비자 지시 사항을 이해한다. -> 기존에 사투리를 듣고 표준어로 변환해 인식하는 것보다 지역별 사투리를 직관적으로 인식하게 된다.
- 인공지능 전화예약 서비스 

## 3. 이미지
- 옷 추천 패션 AI
  - 내 취향을 잘 아는 누군가 알아서 옷을 골라주고 집까지 배달해 주는 것.
  - 소비자가 입력한 데이터를 토대로 의상을 추천해준다.
  - AI를 활용한 알고리즘을 통해 고객 취향과 체형에 맞는 옷을 스타일링
  - 온라인 패션커머스 기업 무신사의 경우 인공지능 기반 이미지 검색 서비스
  - ex. 스티치픽스, 무신사
- 인공지능 '스마트 오프라인' 패션 플랫폼
  - 빅데이터 분석을 통해, 잘 팔리는 아이템을 골라준다.
  - 소비자가 많이 찾는 상품을 공항 패션, 드라마 패션, 동대문 유행 패션, 인기 컬러 등 이미지로 검색해 데이터를 뽑는다.
  - 원피스, 카디건, 재킷 등 키워드 검색이 아닌 구매한 상품의 이미지를 통해 데이터를 도출하기 때문에 적중률이 매우 높다.


## 4. 자율 주행
인지, 판단, 제어
- 인지 : 
  - 차량에 장착된 카메라나 레이더(Radar), 라이다(Lidar)와 같은 센서로 주변환경을 파악하고, 스캐너로 정확한 정보를 습득해 인지
  - 경로 탐색 : ① GPS, 주변 환경을 모니터링하고 장애물을 인지하는 레이저 스캐너, 도로 위의 횡단보도를 인식하는 영상센서가 주요 구성 항목이다.
  - 정확한 정보 수집을 위해 오차범위가 적은 ② DGPS, 보다 정밀한 측정 기술을 구현하기 위한 레이저 스캐너, 스테레오 카메라와 같은 고성능 센서를 집중 개발
  - V2X(Vehicle to Everything)는 주변 환경과의 통신을 통해서 주행하는 기술
- 판단 :
  - 주행상황 판단 및 주행전략 결정
    ex. 차선변경, 추월, 좌/우 회전, 정차 등
  - 주행 경로 생성
- 차량 제어
  - 엔진 가감속, 조향

- 기타
  - 소프트뱅크의 5세대 통신망(5G)를 이용한 군집 주행
  - 자율주행 셔틀 스타트업 ‘드라이브.AI(Drive.AI)’
